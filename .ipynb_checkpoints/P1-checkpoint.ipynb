Below are some helper functions to help get you started. They should look familiar from the lesson!

```python
import math

def grayscale(img):
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    you should call plt.imshow(gray, cmap='gray')"""
    #print("running grayscale()")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #plt.imshow(gray, cmap='gray')
    return gray
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 4 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=5):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    
    imshape = img.shape
    y_max = imshape[0]
    y_min = 330
    
    # vertices for lanes
    left_lane_top_x = 0
    left_lane_top_y = y_min
    left_lane_bottom_x = 0
    left_lane_bottom_y = y_max
    right_lane_top_x = 0
    right_lane_top_y = y_min
    right_lane_bottom_x = 0
    right_lane_bottom_y = y_max
    
    left_lane = []
    right_lane = []
    c_left = 0
    c_right = 0
    m_left = 0
    m_right = 0
    
    # Calculate slopes
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            #print ("left slope: ",slope)
            
            if slope < 0:
                #print("left lane")
                left_lane = line
                m_left = slope
                #print("m_left ", m_left)
                c_left = y2 - (slope * x2)
                break
    
    for line in lines:   
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            #print("right slope: ",slope)
            
            if slope > 0:
                #print("right lane")
                right_lane = line
                m_right = slope
                #print("m_right ", m_right)
                c_right = y2 - (slope * x2)
                break
    
    # Calculate slopes - average method
    c_left_list = []
    c_right_list = []
    m_left_list = []
    m_right_list = []
    point_left_list = []
    point_right_list = []
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2-y1)/(x2-x1)
            if slope<0:
                m_left_list.append(slope)
                c_left_list.append(y2 - (slope * x2))
                point_left_list.append([x1,y1])
            else:
                m_right_list.append(slope)
                c_right_list.append(y2 - (slope * x2))
                point_right_list.append([x1,y1])
    
    # extrapolate
    #print ("m_left, m_right: ", m_left, m_right)
    if not m_left_list or not m_right_list or not c_left_list or not c_right_list:
        return
    
    #print (c_left_list)
    #print (m_left_list)
    #print (sum(c_right_list))
    #print (sum(m_right_list))
    
    c_left_avg = sum(c_left_list)/len(c_left_list)
    c_right_avg = sum(c_right_list)/len(c_right_list)
    m_left_avg = sum(m_left_list)/len(m_left_list)
    m_right_avg = sum(m_right_list)/len(m_right_list)
    
    #print("averages: ", c_left_avg, c_right_avg, m_left_avg, m_right_avg)
    
    if math.isinf(c_left_avg) or math.isinf(c_right_avg) or math.isinf(m_left_avg) or math.isinf(m_right_avg):
        return
    
    left_lane_top_x = int((left_lane_top_y - c_left_avg)/m_left_avg)
    left_lane_bottom_x = int((left_lane_bottom_y - c_left_avg)/m_left_avg)

    right_lane_top_x = int((right_lane_top_y - c_right_avg)/m_right_avg)
    right_lane_bottom_x = int((right_lane_bottom_y - c_right_avg)/m_right_avg)

    #for line in lines:
    #    for x1,y1,x2,y2 in line:
    #        cv2.line(img, (x1, y1), (x2, y2), color, thickness)
    
    #print ("left lane (x1,y1) & (x2,y2): ", left_lane_bottom_x, left_lane_bottom_y, left_lane_top_x, left_lane_top_y)
    #print ("right lane (x1,y1) & (x2,y2): ", right_lane_bottom_x, right_lane_bottom_y, right_lane_top_x, right_lane_top_y)
    
    #print ("Original Lines: ", lines)
    
    new_lines = [[[left_lane_bottom_x, left_lane_bottom_y, left_lane_top_x, left_lane_top_y]], [[right_lane_bottom_x, right_lane_bottom_y, right_lane_top_x, right_lane_top_y]]]
    
    #print ("new Lines: ", new_lines)
    
    for line in new_lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((*img.shape, 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, α=0.8, β=1., λ=0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img * α + img * β + λ
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, α, img, β, λ)
```

```python
import os
import math
os.listdir("test_images/")

for file in os.listdir("test_images/"):
    if file != '.DS_Store':
        # Read each image
        img = mpimg.imread("test_images/" + file)
img = mpimg.imread("test_images/solidYellowLeftTarun.jpg")

# creating a blank to draw lines on
image_copy = np.copy(img)

# Appy grayscale
grayImg = grayscale(img)

#plt.imshow(grayImg, cmap='gray')

# Apply Gaussian Blur
kernel_size = 3
blurGray = gaussian_blur(grayImg, kernel_size)

# Apply Canny
low_threshold = 60
high_threshold = 180
edges = canny(blurGray, low_threshold, high_threshold)

#plt.imshow(edges, cmap='Greys_r')

# masked edges image using cv2.fillPoly()
mask = np.zeros_like(edges)   
ignore_mask_color = 255   

# This time we are defining a four sided polygon to mask
imshape = image.shape
vertices = np.array([[(0,imshape[0]),(430, 330), (550, 330), (imshape[1],imshape[0])]], dtype=np.int32)
cv2.fillPoly(mask, vertices, ignore_mask_color)
masked_edges = cv2.bitwise_and(edges, mask)

#plt.imshow(masked_edges, cmap='Greys_r')

# Apply Hough Transform
rho = 2
theta = np.pi/180
threshold = 10
min_line_len = 55
max_line_gap = 20

lines = hough_lines(masked_edges, rho, theta, threshold, min_line_len, max_line_gap)

#plt.imshow(lines)

# Create a "color" binary image to combine with line image
#color_edges = np.dstack((img, img, img)) 

# Draw the lines on the edge image
lines_edges = weighted_img(lines, image_copy) 

plt.imshow(lines_edges)

# Save the image files (if needed) at this point
#mpimg.imsave("test_images/final_" + file, lines_edges)
#mpimg.imsave("test_images/final_solidYellowLeftTarun.jpg", lines_edges)

print("That's all")
```

### My Thoughts/Reflections

One way to make the algorithm better would be to give a range of angles within which the lanes can fall under, like for left lane, it will be within the 60-80deg range whereas for right lane, between 120-140deg. This way, any lines that show up after hough's algorithm can be filtered based on this criteria.

Another pointer here is that when the lanes curve, the polygon mask area we defined needs to change to avoid the curved lanes inside the hough's algorithm.

Averaging out the slopes will give a smoother lane identification between frames but can also cause the lane to go totally off because of outliers - the slope degree range could solve this issue to remove those outliers.
